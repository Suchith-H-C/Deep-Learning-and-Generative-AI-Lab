```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD, Adam
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split

# Load Iris dataset
iris = datasets.load_iris()
X, y = datasets.load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

# --- Function to build model ---
def build_model(optimizer, loss_function, learning_rate=None):
    model = Sequential()
    model.add(Dense(2, input_shape=(4,)))
    model.add(Activation('relu'))  # ReLU activation
    model.add(Dense(1))
    model.add(Activation('sigmoid'))  # Sigmoid activation

    # If a learning rate is provided, adjust the optimizer
    if optimizer == 'sgd':
        if learning_rate is not None:
            optimizer = SGD(lr=learning_rate)
        else:
            optimizer = SGD()

    model.compile(loss=loss_function, optimizer=optimizer, metrics=['accuracy'])
    return model

# --- (i) Check accuracy using SGD and Adam optimizer ---
optimizers = ['sgd', 'adam']
loss_functions = ['mean_squared_error', 'categorical_crossentropy']

# Check for both optimizers and loss functions
for optimizer in optimizers:
    for loss_function in loss_functions:
        print(f"\n--- Model with {optimizer} optimizer and {loss_function} loss ---")
        model = build_model(optimizer, loss_function)
        model.summary()
        model.fit(X_train, y_train, epochs=300)
        score = model.evaluate(X_test, y_test)
        print(f"Score (Accuracy) for {optimizer} and {loss_function}: {score}")

# --- (ii) Check accuracy by varying learning rates in SGD ---
learning_rates = [0.0001, 0.0005, 5]
for lr in learning_rates:
    print(f"\n--- Model with SGD optimizer and learning rate {lr} ---")
    model = build_model('sgd', 'mean_squared_error', learning_rate=lr)
    model.summary()
    model.fit(X_train, y_train, epochs=300)
    score = model.evaluate(X_test, y_test)
    print(f"Score (Accuracy) for learning rate {lr}: {score}")

```
